```python
import matplotlib.pyplot as plt
import numpy as np

# Визначаємо значення x
x = np.linspace(-10, 10, 400)
# Обчислюємо значення y згідно формули
y = 2 * x + 3

# Створюємо графік
plt.plot(x, y, label='y = 2x + 3')

# Додаємо назву графіка та мітки осей
plt.title('Графік функції y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')

# Додаємо легенду
plt.legend()

# Показуємо графік
plt.show()
```
