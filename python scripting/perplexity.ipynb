Для створення презентації на тему SWOT аналізу у Python ви можете скористатися бібліотекою `matplotlib` для візуалізації діаграм SWOT. Ось приклад коду, який допоможе вам створити просту презентацію SWOT аналізу:

```python
import matplotlib.pyplot as plt

# Дані для SWOT аналізу
labels = ['Сильні сторони', 'Слабкі сторони', 'Можливості', 'Загрози']
sizes = [20, 10, 30, 15]

# Кольори для кожної категорії
colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3']

# Побудова колової діаграми SWOT
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Забезпечує круглу форму діаграми

# Додавання заголовка
plt.title('SWOT Аналіз')

# Показати діаграму
plt.show()
```

Цей код створить просту колову діаграму SWOT аналізу з вказаними вами даними. Ви можете налаштувати розміри секторів, кольори та інші параметри, щоб відобразити ваш аналіз. Також ви можете додати додаткові елементи до презентації, які підкреслять ваші висновки.

````python
import matplotlib.pyplot as plt

# Дані для SWOT аналізу
labels = ['Сильні сторони', 'Слабкі сторони', 'Можливості', 'Загрози']
sizes = [20, 10, 30, 15]

# Кольори для кожної категорії
colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3']

# Побудова колової діаграми SWOT
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Забезпечує круглу форму діаграми

# Додавання заголовка
plt.title('SWOT Аналіз')

# Показати діаграму
plt.show()
````
