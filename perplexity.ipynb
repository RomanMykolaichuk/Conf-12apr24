Штучний інтелект (AI) має значний потенціал для створення навчального контенту, полегшуючи роботу вчителів і роблячи навчання ефективнішим і доступнішим для учнів[1][3]. AI може допомогти вчителям створювати персоналізовані плани навчання, надавати миттєвий зворотний зв'язок, адаптуватися до змін і новацій, і розглядати ці можливості як допоміжний інструмент професійної діяльності[3].

Існує кілька інструментів AI, які можуть бути корисними для створення навчального контенту. Наприклад, IBM Watson Education надає вчителям інструменти для персоналізованого навчання, оцінювання та аналітики[1]. Cognii використовує AI для надання зворотного зв'язку та оцінювання[1]. DeepAI пропонує набір інструментів, що генерують картинки за текстовими запитами[3]. Paintbytext дозволяє редагувати фотографії та створювати матеріали для презентацій за письмовими інструкціями за допомогою ШІ[3]. Bedtimestory використовує AI для створення історій, які читаються дітям перед сном[3].

Крім того, генеративний штучний інтелект значно просунувся вперед, і потреба в різних форматах контенту також зросла[5]. Це означає, що для створення навчального контенту, можна використовувати різні інструменти зі штучним інтелектом, такі як AI-генератор зображень, AI-генератор облич, AI-фотоефект, AI-художній генератор та інші[5].

Однак, при використанні AI для створення навчального контенту, слід враховувати питання конфіденційності та занепокоєння щодо швидкості застосування, масштабів та можливих викликів[4]. Вчителі повинні бути готові до наступу штучного інтелекту та підготувати дітей до майбутнього, якщо теперішнє стрімко змінюється[4].
